# Generate figures for NZMS talk

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

%matplotlib inline
```

# Function to be plotted
We'll use the following function for most of the talk

```python
def f(x, y):
    return 0.5*(1 + np.sin(4*x) * np.cos(7*y)) * 2*np.exp(-10*((x - 0.1)**2 + (y+0.2)**2))
```

We'll generate images of the function both as a surface and as an image. This will be important later

```python
h = 0.1
X, Y = np.meshgrid(np.arange(-1.5, 1.5, h), np.arange(-1.5, 1.5, h))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, f(X, Y), linewidth=0, cmap=cm.gray, rstride=1, cstride=1)
ax.set_xlabel('$x$')
plt.savefig('foo.png', dpi=600)
```
